**Lógica de programação -  Passo a passo para a solução do desafio**

```python
# Passo 0 - Entender o desafio que eu vou resovler;
# Passo 1 - Importar a base de dados (Pasta Vendas);

import pandas as pd
tabela_vendas = pd.read_excel("/content/drive/MyDrive/Python para Ciência de Dados/Vendas/Vendas+-+Base+de+Dados.xlsx")
display(tabela_vendas)

```

```python
# Passo 2 - Calcular o produto mais vendido (em quantidade);
tabela_quantidade_produto = tabela_vendas.groupby('Produto').sum()
tabela_quantidade_produto = tabela_quantidade_produto[['Quantidade']]
tabela_quantidade_produto = tabela_quantidade_produto.sort_values(by='Quantidade', ascending=False)
display(tabela_quantidade_produto)
```

```python
# Passo 3 - Calcular o produto mais vendido (em faturamento);

tabela_vendas['Faturamento'] = tabela_vendas['Quantidade'] * tabela_vendas['Valor Unitário']
tabela_faturamento_produto = tabela_vendas.groupby('Produto').sum()
tabela_faturamento_produto = tabela_faturamento_produto[['Faturamento']].sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento_produto)
```

```python
# Passo 4 - Calcular a loja/estado que mais vendeu (em faturamento);
tabela_faturamento_loja = tabela_vendas.groupby('Loja').sum(numeric_only=True)
tabela_faturamento_loja = tabela_faturamento_loja[['Faturamento']]
tabela_faturamento_loja = tabela_faturamento_loja.sort_values(by='Faturamento', ascending=False)
display(tabela_faturamento_loja)
```

```python
# Passo 5 - Calcular o ticket médio por loja/estado;

tabela_vendas['Ticket Médio'] = tabela_vendas['Valor Unitário']
tabela_ticket_medio = tabela_vendas.groupby('Loja').mean(numeric_only=True)
tabela_ticket_medio = tabela_ticket_medio[['Ticket Médio']]
tabela_ticket_medio = tabela_ticket_medio.sort_values(by='Ticket Médio', ascending=False)
display(tabela_ticket_medio)
```

```python
# Passo 6 - Criar um gráfico/dashboard da loja/estado que mais vendeu (em faturamento);

import plotly.express as px

grafico = px.area(tabela_faturamento_loja, x=tabela_faturamento_loja.index, y='Faturamento')
grafico.show()
```

# **Formatações numéricas e monetárias**

**Loja/Estado que mais vendeu (em faturamento)**

```python
from babel.numbers import format_currency

# Formatação Monetária
tabela_faturamento_loja_formatado = tabela_faturamento_loja['Faturamento'].apply(lambda x: format_currency(x, 'BRL', locale='pt_BR'))
tabela_faturamento_loja_formatado = tabela_faturamento_loja_formatado.reset_index()
display(tabela_faturamento_loja_formatado)
```

**Produtos mais vendidos (em quantidade)**

```python
# Formatação numérica

tabela_quantidade_produto_formatado = pd.DataFrame(tabela_quantidade_produto['Quantidade'].apply(lambda v: "{:,}".format(v).replace(',', '.')))
tabela_quantidade_produto_formatado = tabela_quantidade_produto_formatado.reset_index()
display(tabela_quantidade_produto_formatado)
```

**Produtos mais vendidos (em faturamento)**

```python
# Formatação monetária

tabela_faturamento_produto_formatado = pd.DataFrame(tabela_faturamento_produto['Faturamento'].apply(lambda v: format_currency(v, 'BRL', locale='pt_BR')))
tabela_faturamento_produto_formatado = tabela_faturamento_produto_formatado.reset_index()

display(tabela_faturamento_produto_formatado)
```

**Ticket Médio por Loja/Estado**

```python
# Formatação monetária

tabela_ticket_medio_formatado = pd.DataFrame(tabela_ticket_medio['Ticket Médio'].apply(lambda v: format_currency(v, 'BRL', locale='pt_BR')))
tabela_ticket_medio_formatado = tabela_ticket_medio_formatado.reset_index()
display(tabela_ticket_medio_formatado)
```
